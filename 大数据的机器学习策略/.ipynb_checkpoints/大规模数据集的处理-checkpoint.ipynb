## 大规模数据集中的机器学习

![001](images/001.png)

面对大数据样本，我们通常会寻找更加有效，高效的数据处理算法，如下介绍：

### 1.随机梯度下降

不管是线性回归，逻辑回归，神经网络等模型都会定义一个损失函数，并且使用梯度下降算法进行迭代减小损失函数的值。当面对大数据样本时，这样的方式不再适用。

采用随机梯度下降算法;(左边为批量梯度算法，右边为随机梯度算法)


![002](images/002.png)


+ 1.随机打乱所有的样本顺序；

+ 2.遍历每一个样本更新迭代参数值，降低了计算量。批量梯度下降每次循环都要考虑所有的样本，而随机梯度则循环样本数量。

![003](images/003.png)

如何保证梯度收敛？

批量梯度下降算法（样本容量小时）可以绘制成本函数值，保证每一次都是收敛。

随机梯度算法（样本量非常大时）可以绘制每隔一段数量（如，1000次）来计算一次成本函数（均值），总体收敛即可。



![006](images/006.png)


![007](images/007.png)

+ 1.正常下降，调整学习率即可；

+ 2.每5000个样本求一次均值，会得到更加平滑的线；

+ 3.由于计算成本时迭代次数较小，会出现曲线下降不明显，增加计算均值的样本数量即可；

+ 4.成本函数警告，使用较小的学习率进行学习。

### 3.在线学习机制（随机梯度下降的变种）

当**连续**用户较多时，可以不断的提取用户特征，进行在线训练参数。在线学习机制就是根据新样本不断迭代参数，学习参数的过程。每当一个样本来到时，就要更新参数，让后放弃该样本，等待下一个样本的到来，根据样本的变化，使算法渐渐适应样本的变化。

![009](images/009.png)

在线学习可以学习到用户的变化。前提是有连续的输入数据。
