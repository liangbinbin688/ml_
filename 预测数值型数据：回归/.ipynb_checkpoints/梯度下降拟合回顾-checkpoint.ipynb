梯度下降算法寻找线性拟合参数：

    通过Cost函数对${\theta}$的偏导，各分量都有一个方向导数，最后求一个总的方向下降，目的是为了减少代价函数。其中根据步长去迭代更新目标函数的参数，一直达到一定的目标。
  梯度下降法(gradient  descent)别名最速下降法（曾经我以为这是两个不同的算法-.-）,是用来求解无约束最优化问题的一种常用算法。下面以求解线性回归为题来叙述：
  设一般的线性回归（拟合函数）为：（其中${\theta}_0$的值为1）$$h(x)={\theta}_0b+{\theta}_1x_1+...{\theta}_nx_n$$
    则这组向量参数选择的好坏就需要一个机制来评估，据此我们提出了其损失函数（选择均方误差）：
   $$loss({\theta})=\frac{1}{2m}\sum_{i=1}^{m}(h_{\theta}(x^i)-y^i)^2$$
    
我们现在的目的就是使得损失函数取得最小值。即目标函数为：
        $$min loss({\theta})$$
        
如果上式的值取得0，意味着我们构造了极好的拟合函数，也即选择了最好的${\theta}$的值，但这基本是达不到的，我们只能使得其无限的接近于0，当满足一定精度时停止迭代。

取{\theta}的最优解，方法很多，此处以梯度下降法为例：

+ 初始化${\theta}$的值
+ 改变${\theta}$的值，使得$loss({\theta})$按梯度下降的方向减少。

${\theta}$的更新使用如下的方式来完成：

 $${theta}_i={theta}_i-{\alpha}\frac{loss({\theta})}{{\theta}偏导}$$
 
 其中为步长因子，这里我们取定值，但注意如果取得过小会导致收敛速度过慢，过大则损失函数可能不会收敛，甚至逐渐变大，可以在下述的代码中修改的值来进行验证。后面我会再写一篇关于随机梯度下降法的文章，其实与梯度下降法最大的不同就在于一个求和符号。
