## 异常检测

异常检测就是根据现有数据进行建模，利用现有模型检测新的数据点是否正常，及时提醒。

常用于用户行为检测，等。

如何选择一元的模型和多元的高斯模型？

+ 一元比较常见，计算成本比较低。需要手动的组合特征，

+ 多元模型可以自动的得到特征直接的关系，但是计算成本比较高。需要计算协方差矩阵，成本较高。还要满足样本数量大于特征值的数量。如果发现协方差矩阵是不可逆的，就要检测特征是否有冗余(特征之间存在线性相关)或者样本数量远大于特征值的个数。
